## Investigation Overview

> This project Investigates the influence of the parents background, test preparation etc on students performance

## Dataset Overview

> There are 1,000 students in the dataset with 8 features (gender,race/ethnicity,parental level of education,launch,test prepration course,math score,reading score,writing score, and an addtional created feature overall mean score


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df =pd.read_csv("StudentsPerformance.csv")
df['overall_score']=(df['writing score']+df['reading score']+df['math score'])/3
df['parental level of education']=df['parental level of education'].replace({'some high school':'high school'})
```

## (Visualization 1)

> Students who had the standard lunch had an average overall score greater than those with free/reduced lunch.
and also had a narrower grades distribuation

```python
sns.boxplot(data=df,y='overall_score',x='lunch',color=sns.color_palette()[0])
plt.title('lunch and overall_score')
plt.show()
```

## (Visualization 2)

>  The average math and reading test scores tend to be higher with those who completed a test preparation course.<br>

```python
for item in df.columns[-4:-2]:
    sns.boxplot(x=df["test preparation course"], y=df[item],color=sns.color_palette()[0])
    plt.title(item+" and pre test", loc="left")
    plt.show()
```

## (Visualization 3)
>Math scores and Writing scores have a postive correlation with a distrubation peeks approximately between 60 and 80


```python
x_binss=np.arange(10,100+10,10)
y_bins=np.arange(0,100+10,10)
plt.hist2d(data=df,x='writing score',y='math score',cmin=.5,cmap='viridis_r',bins=[x_binss,y_bins]);
plt.colorbar();
plt.title('writing score and math score heatmap');
plt.xlabel('writing score');
plt.ylabel('math score');
```
